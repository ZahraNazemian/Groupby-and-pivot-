```python
import pandas as pd
import numpy as np
# Step 2: Load the CSV file
df = pd.read_csv('/Users/zahranazemian/Desktop/Python/cars.csv')
df.head(3)
```

```python
df['drive-wheels'].unique()
```

```python
df_group_one = df[['drive-wheels','body-style','price']]
```

```python
df_grouped = df_group_one.groupby(['drive-wheels'], as_index=False).agg({'price': 'mean'})
df_grouped
```

```python
df_grouped1 = df_group_one.groupby(['drive-wheels','body-style'],as_index=False).mean()
df_grouped1
```

```python
grouped_pivot = df_grouped1.pivot(index='drive-wheels',columns='body-style')
grouped_pivot
```
